### 8.Check the outliers and handle outliers in performance score using Percentiles.

```python
plt.boxplot(df3['performance_score'])
plt.title('Box plot of Performance Score');
```

```python
Q1=np.percentile(df3['performance_score'],25,interpolation='midpoint')
Q2=np.percentile(df3['performance_score'],50,interpolation='midpoint')
Q3=np.percentile(df3['performance_score'],75,interpolation='midpoint')
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
low_limit=Q1-1.5*IQR
up_limit=Q3+1.5*IQR
print(low_limit)
print(up_limit)
```

```python
outlier=[]
for x in df3['performance_score']:
    if((x<low_limit)or(x>up_limit)):
        outlier.append(x)
```
